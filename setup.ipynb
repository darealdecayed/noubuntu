```python
#@title Hit RUN to Launch Ubuntu + XFCE + Chrome via noVNC (User: axisvms) 
import os

os.system("""
printf 'Setting up Ubuntu + XFCE + noVNC for user axisvms... '\n
# Create user
sudo useradd -m axisvms
echo 'axisvms:axispass' | sudo chpasswd

# Install minimal desktop + noVNC
sudo apt-get update -qq
sudo apt-get install -y xfce4 xfce4-goodies x11vnc novnc websockify wget curl

# Install Google Chrome
wget -q https://dl.google.com/linux/direct/google-chrome-stable_current_amd64.deb
sudo dpkg -i google-chrome-stable_current_amd64.deb || sudo apt-get install -f -y

# Start virtual X server
sudo -u axisvms Xvfb :1 -screen 0 1024x768x16 &

# Start XFCE session
sudo -u axisvms startxfce4 &

# Start noVNC server on port 8080
sudo -u axisvms websockify -D --web=/usr/share/novnc/ 8080 localhost:5900

printf '\nSetup Complete.\n'
printf 'Open your Binder URL + /proxy/8080 to access the desktop.\n'
""")
```
